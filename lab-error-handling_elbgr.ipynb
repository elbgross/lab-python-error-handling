````python
products=["t-shirt","mug","hat","book","keychain"]

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
inventory = initialize_inventory(products)
print(inventory)
````

````python
def total_price_order(customers_orders):
    price_by_product={}
    for products in customers_orders:
        valid_price=False
        while not valid_price:
            try:
                price=float(input(f"Enter the price for {products}: "))
                if price<0:
                    raise ValueError("Invalid price! Please enter a non-negative value.")
                valid_price=True
            except ValueError as error:
                print(f"Error: {error}")
        price_by_product[products]=price
    return price_by_product
        
price_by_product= total_price_order(customers_orders)
print(price_by_product)
````
